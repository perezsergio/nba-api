```python
import pandas as pd

regseason = pd.read_csv("../data/regseason_yearly_stats.csv")
regseason.head()
```

```python
# Compute total number of points per season
regseason["TOTAL_PTS"] = regseason["GP"] * regseason["PTS"]

# sort by total points
regseason_sorted = regseason.sort_values(by="TOTAL_PTS", ascending=False)

# Get the season with the highest total points
row_max_points = regseason_sorted.head(1)
season_max_points = row_max_points["SEASON_ID"].values[0]
print(f"The season with the highest total points is {season_max_points}")

# Get the season with the lowest total points
row_min_points = regseason_sorted.tail(1)
season_min_points = row_min_points["SEASON_ID"].values[0]
print(f"The season with the highest total points is {season_min_points}")
```

```python
total_career_points = regseason["TOTAL_PTS"].sum()
total_career_games = regseason["GP"].sum()
avg_career_points = total_career_points / total_career_games
print(
    f"The points per game, averaged throughout his whole career, is: {round(avg_career_points,2)} pts."
)
```

```python
regseason["TOTAL_REB"] = regseason["GP"] * regseason["REB"]
total_career_rebounds = regseason["TOTAL_REB"].sum()
avg_career_rebounds = total_career_rebounds / total_career_games
print(
    f"The rebounds per game, averaged throughout his whole career, is: {round(avg_career_rebounds,2)} reb."
)
```

```python
regseason["TOTAL_AST"] = regseason["GP"] * regseason["AST"]
total_career_assists = regseason["TOTAL_AST"].sum()
avg_career_assists = total_career_assists / total_career_games
print(
    f"The assists per game, averaged throughout his whole career, is: {round(avg_career_assists,2)} ast."
)
```
